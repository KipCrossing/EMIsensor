```python
import csv
import numpy as np
import matplotlib.pyplot as plt

plot_vector2 = []

with open('2Dfield/0m/([1, 0, 1], [-1, 0, 1]).csv') as csvfile:
    readCSV = csv.reader(csvfile, delimiter=',')
    for row in readCSV:
        try:
            row_float = [float(i) for i in row]
            plot_vector2.append(row_float)
        except:
            pass
    plot_vector2.pop(0)



soa =np.array(plot_vector2) 
X,Y,U,V = zip(*soa)
plt.figure()
ax = plt.gca()
ax.quiver(X,Y,U,V,angles='xy',scale_units='xy',scale=10000000000000)
ax.set_xlim([-1,3])
ax.set_ylim([-2,1])
plt.draw()
plt.show()
```
