```python
import csv
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

path = "C:/Users/kcro5515/EMIsensor/"

file = "3D_Field_0.2_([0, 0, 1], [0, 0, 0])_None_([0, 0, 1], [1, 0, 0]).csv"

plot_vector2 = []

with open(path+file) as csvfile:
    readCSV = csv.reader(csvfile, delimiter=',')
    for row in readCSV:
        try:
            row_float = [float(i) for i in row]
            plot_vector2.append(row_float)
        except:
            pass
    plot_vector2.pop(0)



soa =np.array(plot_vector2) 
X,Y,Z,U,V,W = zip(*soa)
fig = plt.figure()
ax = fig.gca(projection='3d')
ax.set_xlim3d(-1, 2)
ax.set_ylim3d(-1, 1)
ax.set_zlim3d(-1, 0)
ax.quiver(X,Y,Z,U,V,W,length = 0.2)

#plt.draw()
plt.show()
```

```python
plt.figure()
axi = plt.gca()
axi.quiver(X,Z,U,V,angles='xy',scale_units='xy',scale=0.01)
axi.set_xlim([-1,3])
axi.set_ylim([-2,1])
plt.draw()
plt.show()
```
